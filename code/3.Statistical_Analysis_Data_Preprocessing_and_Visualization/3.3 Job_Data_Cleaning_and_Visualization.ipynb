This part of my project involves crucial steps of data cleaning and advanced visualization focusing on wage data across different job categories. Here's an overview of what each part of the script accomplishes:

1. **Data Reading**:
   - The script begins by loading job wage data from a JSON file into a pandas DataFrame. This is the initial step to manipulate and visualize the data.

2. **Data Cleaning**:
   - identify key columns related to wage percentiles (10th, 25th, 50th, 75th, and 90th) and convert them to numeric types, handling non-numeric values with coercion to `NaN` (i.e., treating any errors by setting invalid parsing as `NaN`).
   - Following the conversion, drop all rows where any of the wage percentile columns contain `NaN`, ensuring that only complete and clean data is used for analysis.

3. **Data Visualization**:
   - **Box Plot**: A box plot is generated to visualize the distribution of wages across different percentiles for various jobs. This visualization helps identify outliers and provides a clear summary of the wage distribution characteristics, such as the median and quartiles.
   - **Scatter Plot Matrix**: A scatter plot matrix (or pair plot) is used to observe pairwise relationships between the different wage percentiles. This helps in identifying patterns, correlations, or potential anomalies between different wage levels.

4. **Data Saving**:
   - After cleaning, the refined dataset is saved back into a JSON file, preserving the cleaned state for future use or further analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Read Data
file_path = '/Users/a1234/Desktop/workspace/Employment_Analysis_and_Recommendation_System_Based_on_NLP_and_Data_Modeling/data/Job_Data_Sum.json'
data = pd.read_json(file_path)

# Step 2: Data Cleaning
cols_to_check = ['Annual Wage 10', 'Annual Wage 25', 'Annual Wage 50', 'Annual Wage 75', 'Annual Wage 90']
for col in cols_to_check:
    data[col] = pd.to_numeric(data[col], errors='coerce')
data_cleaned = data.dropna(subset=cols_to_check)

# Step 3: Data Visualization

# Box plot to see the wage distribution for different jobs
plt.figure(figsize=(12, 8))
data_cleaned[cols_to_check].plot(kind='box', vert=False)
plt.title('Salary Distribution by Percentile')
plt.xlabel('Annual Wage')
plt.ylabel('Percentiles')
plt.grid(True)
plt.show()

# Scatter Plot Matrix
# Generating scatter plot matrix to visualize relationships between salary percentiles
sns.pairplot(data_cleaned[cols_to_check])
plt.suptitle('Scatter Plot Matrix of Salary Percentiles', y=1.02)
plt.show()

# Step 4: Save Cleaned Data
output_file_path = '/Users/a1234/Desktop/workspace/Employment_Analysis_and_Recommendation_System_Based_on_NLP_and_Data_Modeling/data/Cleaned_Job_Data.json'
data_cleaned.to_json(output_file_path, orient='records')

print("Data cleaning completed, and advanced visualizations have been performed. Results have been saved to a new JSON file.")
```
